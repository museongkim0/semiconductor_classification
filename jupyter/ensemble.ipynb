```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(n_samples=500, noise=0.3, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
bag_clf = BaggingClassifier(DecisionTreeClassifier(), n_estimators=500, max_samples=100, bootstrap=True)
```

```python
bag_clf.fit(X_train,y_train)
```

```python
y_pred=bag_clf.predict(X_test)
```

```python
y_pred
```

```python
bag_clf = BaggingClassifier(DecisionTreeClassifier(), n_estimators=500, max_samples=100, bootstrap=True,oob_score=True)
```

```python
bag_clf.fit(X_train,y_train)
```

```python
bag_clf.oob_score_
```

```python
from sklearn.metrics import accuracy_score
y_pred = bag_clf.predict(X_test)
accuracy_score(y_test,y_pred)
```

## GBM

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_reg1 = DecisionTreeRegressor(max_depth = 2) # weak learner
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
tree_reg1.fit(X,y)
```

```python
y2 = y-tree_reg1.predict(X) # y -> residual
```

```python
tree_reg2 = DecisionTreeRegressor(max_depth = 2)
tree_reg2.fit(X,y2)
```

```python
y3 = y2-tree_reg2.predict(X) # y2 -> residual
```

```python
tree_reg3 = DecisionTreeRegressor(max_depth = 2)
tree_reg3.fit(X,y3)
```

```python
y_pred = sum(tree.predict([X[1]]) for tree in (tree_reg1,tree_reg2,tree_reg3))
```

```python
y_pred
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=3,learning_rate=1)
```

```python
gbrt.fit(X,y)
```

```python
gbrt.predict([X[1]])
```

```python
ada_clf.fit(X_train, y_train)
```

```python
ada_clf.predict(X_test)
```
